```python
from __future__ import print_function

import pylab as pl
# this package sets up pretty plots
# import seaborn
from scipy.optimize import curve_fit, minimize
    
%pylab inline

import numpy as np

```

### Assignment 1 _ HW 3 : Chi Squared, Normal, Poisson, Binomial and Uniform distributions

```python
mean = 100
size = 50
mysize = (2000 / (np.array(range(1, 100)))).astype(int)

md = {}
```

### Here is the Chi Square part:

```python
# chisq df = mean
df = mean

md['chisq'] = np.random.chisquare(df, size)
pl.figure(figsize=(3, 3))
pl.hist(md['chisq'], bins = 15)
pl.ylabel('N')
pl.xlabel('x')

print ("Chi Square mean: %.2f, standard deviation: %.2f"%(md['chisq'].mean(), md['chisq'].std()))

```

```python


md['chisq'] = {} 

for n in mysize:
    md['chisq'][n] = np.random.chisquare(df, size = n)
    
# save the means for plotting later    

md['chisq']['means'] = {}
axchisq_mu_n = pl.figure(figsize=(5,3)).add_subplot(111)

for nn in md['chisq']:
    if not type(nn) == str:
        md['chisq']['means'][nn] = md['chisq'][nn].mean()
        #and plot it
        axchisq_mu_n.plot(nn, md['chisq']['means'][nn], 'o')
        axchisq_mu_n.set_xlabel('sample size', fontsize=18)
        axchisq_mu_n.set_ylabel('sample mean', fontsize=18)
        axchisq_mu_n.set_title('Chi squared', fontsize=18)
        axchisq_mu_n.plot([min(mysize), max(mysize)], [df, df], 'k')
    
```

```python
allmeans = list(md['chisq']['means'].values())

pl.figure(figsize=(3, 3))
pl.hist(allmeans,bins=15)
pl.xlabel('sample mean - Chi Sq', fontsize = 18)
pl.ylabel('N', fontsize = 18)
```

### Here is the Normal Dist:

```python
# for normal samples
sd = 10

md['norm'] = np.random.normal(mean, sd, size)

pl.figure(figsize=(3, 3))
pl.hist(md['norm'], bins = 15)
pl.ylabel('N')
pl.xlabel('x')

print ("Norma' Dist mean: %.2f, standard deviation: %.2f"%(md['norm'].mean(), md['norm'].std()))
```

```python
md['norm'] = {}

for n in mysize:
    md['norm'][n] = np.random.normal(mean, sd, size=n)
    
# save the means for plotting later    

md['norm']['means'] = {}
norm_plot = pl.figure(figsize=(5,3)).add_subplot(111)

for nn in md['norm']:
    if not type(nn) == str:
        md['norm']['means'][nn] = md['norm'][nn].mean()
        #and plot it
        norm_plot.plot(nn, md['norm']['means'][nn], 'o')
        norm_plot.set_xlabel('sample size', fontsize=18)
        norm_plot.set_ylabel('sample mean', fontsize=18)
        norm_plot.set_title('Normal Dist', fontsize=18)
        norm_plot.plot([min(mysize), max(mysize)], [mean, mean], 'k')
```

```python
allmeans = list(md['norm']['means'].values())

pl.figure(figsize=(3, 3))
pl.hist(allmeans,bins=15)
pl.xlabel('sample mean - normal', fontsize = 18)
pl.ylabel('N', fontsize = 18)
```

### Here is the binomial distribution:

```python
# for binomial 
trials = 200
prob = .5

md['bino'] = np.random.binomial(trials, prob, size)

pl.figure(figsize=(3, 3))
pl.hist(md['bino'], bins = 15)
pl.ylabel('N')
pl.xlabel('x')
print ("Binomial mean: %.2f, standard deviation: %.2f"%(md['bino'].mean(), md['bino'].std()))
```

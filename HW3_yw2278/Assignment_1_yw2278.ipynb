```python
import numpy as np
import matplotlib.pyplot as plt
# this package sets up pretty plots
from scipy.optimize import curve_fit, minimize
from scipy import stats  
%matplotlib inline
```

### Gaussian Distribution


```python
#distributions = ['pois', 'gaus', 'chisq', 'cauchy', 'lnorm', 'binomial']
distributions = ['gaus']
```

```python
# for convinience, generating 199 samples with size ranging from 10 to 20000
mymean = 10
mysize = (np.arange(10,2000,10)).astype(int)
```

```python
md = {}
# md is an empty disctionary

md['gaus'] = 2*np.random.randn(1000)+mymean
pl.hist(md['gaus'], bins = 30)
pl.ylabel('N', fontsize = 18)
pl.xlabel('x', fontsize = 18)
pl.title("Gaussian Distribution", fontsize = 18)
```

```python
mdis={}
mdis["gaus"]={}
```

```python
mdis['gaus']["value"] = [np.random.randn(i)+mymean for i in mysize]
```

```python
mdis['gaus']['means'] = {}
ax_mu_n = pl.figure(figsize=(10,6)).add_subplot(111)
for nn in range(np.size(mdis['gaus']["value"])):
    mdis['gaus']['means'][nn] = mdis['gaus']["value"][nn].mean()
    ax_mu_n.plot(np.size(mdis['gaus']["value"][nn]), mdis['gaus']['means'][nn], 'o')
    ax_mu_n.set_xlabel('sample size', fontsize=18)
    ax_mu_n.set_ylabel('sample mean', fontsize=18)
    ax_mu_n.set_title('Gaussain', fontsize=18)
    ax_mu_n.plot([min(mysize), max(mysize)], [mymean, mymean], 'k')  
```

```python
#define a gaussian function
def gaufunc(x, miu, sigma, t):
    a = -(x-miu)**2/(2*sigma**2)
    return t*exp(a)
```

```python
allmeans = list(mdis['gaus']['means'].values())
pl.figure(figsize = (10, 10))
(n,bins,patches) = pl.hist(allmeans,bins = 50,alpha = 0.2)
x_lin = np.linspace(bins[0], bins[-1], np.size(n))
x_space = np.linspace(bins[0], bins[-1], 10000)
popt, pcov = curve_fit(gaufunc, x_lin, n, p0 = [10, 1, 10])
y = gaufunc(x_space, *popt)
pl.xlabel('sample mean', fontsize = 18)
pl.ylabel('N', fontsize = 18)
pl.plot(x_space, y, label= 'fit')
pl.legend(loc = "upper left")
pl.title('Histogram and fit curve',fontsize = 18)
```
